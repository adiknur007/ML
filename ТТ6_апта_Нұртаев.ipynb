```
# 4-Тапсырма
dt = {
    "drug": ["A", "B", "C", "D"],
    "price": [1200, 800, 600, 2000],
    "amount": [5, 10, 3, 8]
}
df = pd.DataFrame(dt)
print(df)

a = df[df["price"]>1000]
print (a)

b = df[(df["price"] >= 600) & (df["price"] <= 800)]
print (b)

c = df[df["amount"]<5]
print (c)
```

```
dt = {
    "name": ["Ayan", "Dana", "Miras", "Aida"],
    "departament": ["IT", "HR", "IT", "HR"],
    "salary": [400000, 300000, 450000, 320000]
}
df = pd.DataFrame(dt)
print(df)

ortasha = df.groupby("departament")["salary"].mean()
print (ortasha)

sany = df.groupby("departament")["name"].count()
print(sany)

en_zhogary = ortasha.idxmax()
print(en_zhogary)
```
